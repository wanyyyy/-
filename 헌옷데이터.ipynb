## 지오코딩

```python
key = '시크릿'
address = ch['도로명 주소']
```

```python
data_ch= {}
for name in address:
    try : 
        r = f'http://api.vworld.kr/req/address?service=address&request=getCoord&key={key}&address={name}&type=road'
        res=requests.get(r)
        x=res.json()['response']['result']['point']['x']
        y=res.json()['response']['result']['point']['y']
        data_ch[name] = [x,y]
    except:
        continue
```

```python
data_ch
```

```python
df_ch = pd.DataFrame(data_ch)
df_ch
```

```python
df_ch = df_ch.transpose()
df_ch = df_ch.reset_index()
df_ch.columns=['주소','경도','위도']
df_ch
```

# 오픈API로 도로명주소, 지번주소 가져오기

```python
df['point'] = df['경도']+','+df['위도']
df
```

```python
point = df['point']
```

```python
juso1= {}
juso2= {}
for i in point:
    try : 
        r = f'http://api.vworld.kr/req/address?service=address&request=getAddress&key={key}&point={i}&refine=false&type=both'
        res=requests.get(r)
        a = res.json()['response']['result'][0]['text']
        b = res.json()['response']['result'][1]['text']
        juso1[i] = [a]
        juso2[i] = [b]
    except:
        continue
```

```python
juso1
```

```python
juso2
```

```python
juso_1 = pd.DataFrame(juso1)
juso_1 = juso_1.transpose()
juso_1 = juso_1.reset_index()
juso_2 = pd.DataFrame(juso2)
juso_2 = juso_2.transpose()
juso_2 = juso_2.reset_index()
juso = pd.concat([juso_1.iloc[:], juso_2.iloc[:,1]], axis=1)
juso
```

```python
juso.columns=['point','지번주소','도로명주소']
juso
```
